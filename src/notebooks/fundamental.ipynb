```python
def to_float(df, cols):
    return df.with_columns([
        pl.col(c).replace("None", None).cast(pl.Float64)
        for c in cols
    ])
```

```python
income = to_float(income, ["totalRevenue", "grossProfit", "netIncome"])
balance = to_float(balance, ["totalLiabilities", "totalShareholderEquity"])
```

```python
income = income.sort(["symbol", "fiscalDateEnding"])
balance = balance.sort(["symbol", "fiscalDateEnding"])
cashflow = cashflow.sort(["symbol", "fiscalDateEnding"])
```

```python
income = income.with_columns([
    pl.col("fiscalDateEnding").cast(pl.Date)
])

balance = balance.with_columns([
    pl.col("fiscalDateEnding").cast(pl.Date)
])

cashflow = cashflow.with_columns([
    pl.col("fiscalDateEnding").cast(pl.Date)
])
```

```python
# Income based features

income = income.with_columns([
    (pl.col("grossProfit") / pl.col("totalRevenue")).alias("gross_margin")
])
```

```python
# Balancesheet features

balance = balance.with_columns([
    (pl.col("totalLiabilities") / pl.col("totalShareholderEquity"))
    .alias("debt_to_equity")
])
```

```python
# Cashflow features

cashflow = cashflow.with_columns([
    (
        pl.col("operatingCashflow") -
        pl.col("capitalExpenditures").abs()
    ).alias("free_cash_flow")
])
```

```python
# Merge income + balance sheet

df = income.join(
    balance.select([
        "symbol",
        "fiscalDateEnding",
        "totalShareholderEquity",
        "debt_to_equity"
    ]),
    on=["symbol", "fiscalDateEnding"],
    how="left"
)
```

```python
# Add cashflow data

df = df.join(
    cashflow.select([
        "symbol",
        "fiscalDateEnding",
        "operatingCashflow",
        "capitalExpenditures"
    ]),
    on=["symbol", "fiscalDateEnding"],
    how="left"
)
```

```python
# Computing cashflow features after join

df = df.with_columns([
    (
        pl.col("operatingCashflow") -
        pl.col("capitalExpenditures").abs()
    ).alias("free_cash_flow")
])
```

```python
# ROE

df = df.with_columns([
    (
        pl.col("netIncome") /
        pl.col("totalShareholderEquity")
    ).alias("roe")
])
```

```python
# FCF Margin

df = df.with_columns([
    (
        pl.col("free_cash_flow") /
        pl.col("totalRevenue")
    ).alias("fcf_margin")
])
```

```python
# Revenue growth QoQ

df = df.sort(["symbol", "fiscalDateEnding"])

df = df.with_columns([
    (
        pl.col("totalRevenue") /
        pl.col("totalRevenue").shift(1).over("symbol") - 1
    ).alias("revenue_growth_qoq")
])
```

```python
# Gross Margin + QoQ

df = df.with_columns([
    (pl.col("grossProfit") / pl.col("totalRevenue")).alias("gross_margin")
])

df = df.with_columns([
    (
        pl.col("gross_margin") -
        pl.col("gross_margin").shift(1).over("symbol")
    ).alias("gross_margin_qoq")
])
```

```python
df = df.with_columns([
    pl.col("revenue_growth_qoq").clip(-5, 5),
    pl.col("fcf_margin").clip(-5, 5),
    pl.col("gross_margin_qoq").clip(-1, 1)
])
```

```python
def winsorize(col, lower=-3, upper=3):
    return pl.col(col).clip(lower, upper)

df = df.with_columns([
    winsorize("roe"),
    winsorize("debt_to_equity"),
    winsorize("fcf_margin")
])
```

```python
df = df.with_columns([
    pl.when(pl.col("totalRevenue") == 0)
    .then(None)
    .otherwise(pl.col("free_cash_flow") / pl.col("totalRevenue"))
    .alias("fcf_margin")
])
```

```python
# Final feature table

final = df.select([
    "symbol",
    "fiscalDateEnding",
    "revenue_growth_qoq",
    "gross_margin",
    "gross_margin_qoq",
    "debt_to_equity",
    "fcf_margin",
    "roe"
])
```
