```python
def to_float(df, cols):
    return df.with_columns([
        pl.col(c).replace("None", None).cast(pl.Float64)
        for c in cols
    ])
```

```python
income = to_float(income, ["totalRevenue", "grossProfit", "netIncome"])
balance = to_float(balance, ["totalLiabilities", "totalShareholderEquity"])
```

```python
income = income.sort(["symbol", "fiscalDateEnding"])
balance = balance.sort(["symbol", "fiscalDateEnding"])
cashflow = cashflow.sort(["symbol", "fiscalDateEnding"])
```

```python
income = income.with_columns([
    pl.col("fiscalDateEnding").cast(pl.Date)
])

balance = balance.with_columns([
    pl.col("fiscalDateEnding").cast(pl.Date)
])

cashflow = cashflow.with_columns([
    pl.col("fiscalDateEnding").cast(pl.Date)
])
```

```python
# Income based features

income = income.with_columns([
    (pl.col("grossProfit") / pl.col("totalRevenue")).alias("gross_margin")
])
```

```python
# Balancesheet features

balance = balance.with_columns([
    (pl.col("totalLiabilities") / pl.col("totalShareholderEquity"))
    .alias("debt_to_equity")
])
```

```python
# Cashflow features

cashflow = cashflow.with_columns([
    (
        pl.col("operatingCashflow") -
        pl.col("capitalExpenditures").abs()
    ).alias("free_cash_flow")
])
```

```python
# Merge income + balance sheet

df = income.join(
    balance.select([
        "symbol",
        "fiscalDateEnding",
        "totalShareholderEquity",
        "debt_to_equity"
    ]),
    on=["symbol", "fiscalDateEnding"],
    how="left"
)
```

```python
# Add cashflow data

df = df.join(
    cashflow.select([
        "symbol",
        "fiscalDateEnding",
        "operatingCashflow",
        "capitalExpenditures"
    ]),
    on=["symbol", "fiscalDateEnding"],
    how="left"
)
```

```python
# Computing cashflow features after join

df = df.with_columns([
    (
        pl.col("operatingCashflow") -
        pl.col("capitalExpenditures").abs()
    ).alias("free_cash_flow")
])
```

```python
# ROE

df = df.with_columns([
    (
        pl.col("netIncome") /
        pl.col("totalShareholderEquity")
    ).alias("roe")
])
```

```python
# FCF Margin

df = df.with_columns([
    (
        pl.col("free_cash_flow") /
        pl.col("totalRevenue")
    ).alias("fcf_margin")
])
```

```python
# Revenue growth QoQ

df = df.sort(["symbol", "fiscalDateEnding"])

df = df.with_columns([
    (
        pl.col("totalRevenue") /
        pl.col("totalRevenue").shift(1).over("symbol") - 1
    ).alias("revenue_growth_qoq")
])
```

```python
# Gross Margin + QoQ

df = df.with_columns([
    (pl.col("grossProfit") / pl.col("totalRevenue")).alias("gross_margin")
])

df = df.with_columns([
    (
        pl.col("gross_margin") -
        pl.col("gross_margin").shift(1).over("symbol")
    ).alias("gross_margin_qoq")
])
```

```python
# QoQ percentage change for debt_to_equity, fcf_margin, roe
df = df.sort(["symbol", "fiscalDateEnding"])

df = df.with_columns([
    (
        pl.col("debt_to_equity") - 
        pl.col("debt_to_equity").shift(1).over("symbol")
    ).alias("debt_to_equity_qoq"),
    
    (
        pl.col("fcf_margin") - 
        pl.col("fcf_margin").shift(1).over("symbol")
    ).alias("fcf_margin_qoq"),
    
    (
        pl.col("roe") - 
        pl.col("roe").shift(1).over("symbol")
    ).alias("roe_qoq")
])
```

```python
df = df.with_columns([
    pl.col("revenue_growth_qoq").clip(-5, 5),
    pl.col("fcf_margin").clip(-5, 5),
    pl.col("gross_margin_qoq").clip(-1, 1)
])
```

```python
def winsorize(col, lower=-3, upper=3):
    return pl.col(col).clip(lower, upper)

df = df.with_columns([
    winsorize("roe"),
    winsorize("debt_to_equity"),
    winsorize("fcf_margin")
])
```

```python
# Fix infinites across all float columns
num_cols = [c for c, t in df.schema.items() if t == pl.Float64]

df = df.with_columns([
    pl.when(pl.col(c).is_infinite())
    .then(None)
    .otherwise(pl.col(c))
    .alias(c)
    for c in num_cols
])

# Winsorize all problematic columns
clip_rules = {
    "gross_margin":       (-1, 1),
    "gross_margin_qoq":   (-1, 1),
    "debt_to_equity_qoq": (-3, 3),
    "fcf_margin":         (-5, 5),
    "fcf_margin_qoq":     (-5, 5),
    "roe_qoq":            (-3, 3),
}

df = df.with_columns([
    pl.col(col).clip(lo, hi).alias(col)
    for col, (lo, hi) in clip_rules.items()
])
```

```python
# Final feature table

final = df.select([
    "symbol",
    "fiscalDateEnding",
    "revenue_growth_qoq",
    "gross_margin",
    "gross_margin_qoq",
    "debt_to_equity",
    "debt_to_equity_qoq",
    "fcf_margin",
    "fcf_margin_qoq",
    "roe",
    "roe_qoq"
])
```

```python
earnings = pl.read_parquet("../ingestion/data/backup/earnings_delta_backup.parquet")

earnings = earnings.with_columns([
    pl.col("reportedEPS").replace("None", None).cast(pl.Float64),
    pl.col("estimatedEPS").replace("None", None).cast(pl.Float64),
    pl.col("surprisePercentage").replace("None", None).cast(pl.Float64),
    pl.col("surprise").replace("None", None).cast(pl.Float64)
])

earnings = earnings.sort(["symbol", "fiscalDateEnding"])

earnings = earnings.with_columns([
    (
        (pl.col("reportedEPS") - pl.col("reportedEPS").shift(1).over("symbol")) /
        pl.col("reportedEPS").shift(1).over("symbol").abs()
    ).alias("eps_growth_qoq")
])

# Fix eps_growth_qoq infinites and clip HERE before join
earnings = earnings.with_columns([
    pl.when(pl.col("eps_growth_qoq").is_infinite())
    .then(None)
    .otherwise(pl.col("eps_growth_qoq"))
    .alias("eps_growth_qoq")
]).with_columns([
    pl.col("eps_growth_qoq").clip(-5, 5)
])
```

```python
# Check if fiscalDateEnding dates match between earnings and final
earnings_dates = earnings.select(["symbol", "fiscalDateEnding"]).sort(["symbol", "fiscalDateEnding"])
final_dates = final.select(["symbol", "fiscalDateEnding"]).sort(["symbol", "fiscalDateEnding"])

# How many rows in each
print(f"Earnings rows: {earnings_dates.shape[0]}")
print(f"Final rows: {final_dates.shape[0]}")

# How many (symbol, fiscalDateEnding) pairs exist in earnings but NOT in final
in_earnings_not_final = earnings_dates.join(final_dates, on=["symbol", "fiscalDateEnding"], how="anti")
print(f"\nIn earnings but NOT in final: {in_earnings_not_final.shape[0]}")
print(in_earnings_not_final.head(10))

# How many exist in final but NOT in earnings
in_final_not_earnings = final_dates.join(earnings_dates, on=["symbol", "fiscalDateEnding"], how="anti")
print(f"\nIn final but NOT in earnings: {in_final_not_earnings.shape[0]}")
print(in_final_not_earnings.head(10))

# How many match exactly
matched = earnings_dates.join(final_dates, on=["symbol", "fiscalDateEnding"], how="inner")
print(f"\nExact matches: {matched.shape[0]}")
```

```python
modeling_fundamentals = earnings.join(
    final,
    on=["symbol", "fiscalDateEnding"],
    how="left"
).select([
    "symbol", "fiscalDateEnding", "reportedDate", "av_quarter",
    "reportedEPS", "estimatedEPS", "surprise", "surprisePercentage",
    "eps_growth_qoq",
    "revenue_growth_qoq",
    "gross_margin", "gross_margin_qoq",
    "debt_to_equity", "debt_to_equity_qoq",
    "fcf_margin", "fcf_margin_qoq",
    "roe", "roe_qoq"
])

# Validate before saving
num_cols = [c for c, t in modeling_fundamentals.schema.items() if t == pl.Float64]
print("=== INFINITE CHECK ===")
for col in num_cols:
    inf_count = modeling_fundamentals.filter(pl.col(col).is_infinite()).shape[0]
    if inf_count > 0:
        print(f"  {col}: {inf_count} infinites — FIX BEFORE SAVING")
    
print("=== EXTREME VALUES ===")
for col in num_cols:
    print(f"{col}: min={modeling_fundamentals[col].min():.4f}, max={modeling_fundamentals[col].max():.4f}")

print(modeling_fundamentals.shape)

print("\n=== NULL COUNTS ===")
print(modeling_fundamentals.null_count())

print(modeling_fundamentals.head())

# Only save if no infinites
modeling_fundamentals.write_parquet(
    "../ingestion/data/fundamentalIndicators/modeling_fundamentals.parquet"
)
print("Saved.")
```
